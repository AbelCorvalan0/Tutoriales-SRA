### Carga de variables generales del circuito.

```python
import sympy as sym
from IPython.display import display, Latex
# Cargo las variables del circuito. 
V1, V2= sym.symbols('V_{1}, V_{2}')
R1, R2, R3, R4, R5= sym.symbols('R_{1}, R_{2}, R_{3}, R_{4}, R_{5}')
```

### Análisis de $V_{1}\neq0V$ y $V_{2}=0V$

Se calcula en primer lugar la ganancia $V_{o1}/V_{1}$

```python
# Cargo las variables particulares del caso de estudio.
Rp, Ir2, Irp, Vo1= sym.symbols('R_{p}, I_{R2}, I_{Rp}, V_{o1}')

#s_Vo1, sVo1= sym.symbols('s_Vo1, sVo1')

equ_Ir2= sym.Eq(Ir2, (Vo1-V1)/R2)
equ_Irp= sym.Eq(Irp, V1/Rp)

# Igualo las ecuaciones
print('Se igualan las ecuaciones')
equ1= sym.Eq(equ_Ir2.rhs, equ_Irp.rhs)
sym.pprint(equ1)

# Resuelvo sistema, se obtiene solucion en formato matriz.
s_Vo1= sym.solve(equ1, Vo1)
sym.pprint(s_Vo1[0])
# Obtengo Vo1
print('Solución: ')
sVo1= sym.Eq(Vo1/V1, s_Vo1[0]/V1)
sym.pprint(sVo1)

# Resultado con Rp
print('Reemplazando Rp se tiene: ')
s_Vo1= sVo1.subs(Rp, (R1*R4)/(R1+R4))
sym.pprint(s_Vo1)

# Fórmula en formato latex
print('Fórmula en formato LaTex: ')
sym.print_latex(s_Vo1)

print('Fórmula con Rp: ')
s_Vo1
```

Se calcula la ganancia $V_{o2}/V_{o1}$. 

```python
# IMPORTANTE: equ_Ir2 funciona también para este panel de código
# si se corre con el botón "Run all", de otra forma el panel puede
# tomar otro valor que no sea el deseado.

# Cargo las variables particulares del caso de estudio.
Ir3, Ir5= sym.symbols('I_{R3}, I_{R5}')
Vo2= sym.symbols('V_{o2}')
s_Vo2, sVo2= sym.symbols('s_Vo2, sVo2')

equ_Ir3= sym.Eq(Ir3, Vo2/R3)
equ_Ir5= sym.Eq(Ir5, -Vo1/R5)

# Igualo las expresiones Ir3 e Ir5.
equ2= sym.Eq(equ_Ir3.rhs, equ_Ir5.rhs)
sym.pprint(equ2)

# Resuelvo sistema, se obtiene solucion en formato matriz.
s_Vo2= sym.solve(equ2, Vo2)

# Obtengo solución
sVo2= sym.Eq(Vo2/Vo1, s_Vo2[0]/Vo1)
sym.pprint(sVo2)

# Fórmula en formato latex
print('Fórmula en formato LaTex: ')
sym.print_latex(sVo2)

sVo2
```

Ganancia $A_{V1}= \frac{V_{o2}}{V_{o1}} \frac{V_{o1}}{V_{1}}$.

```python
Av1, R= sym.symbols('A_{V1}, R')

equ_Av1= sym.Eq(Av1, sVo2.rhs*s_Vo1.rhs)

sym.pprint(equ_Av1)
sym.print_latex(sym.simplify(equ_Av1))
equ_Av1.subs({R1: R, R2: R, R3: R, R4: R, R5: R})
```

### Análisis de $V_{1}=0V$ y $V_{2}\neq0V$

Se calcula en primer lugar la ganancia $V_{o1}/V_{2}$

```python
# Cargo las variables particulares del caso de estudio.
Ir1, R= sym.symbols('I_{R1}, R')

# Escribo las ecuaciones de Ir1 e Ir2.
equ_Ir3= sym.Eq(Ir2, Vo1/R2)
equ_Ir5= sym.Eq(Ir1, -V2/R1)

# Igualo las expresiones Ir3 e Ir5.
equ2= sym.Eq(equ_Ir3.rhs, equ_Ir5.rhs)
sym.pprint(equ2)

# Resuelvo sistema, se obtiene solucion en formato matriz.
s_Vo1= sym.solve(equ2, Vo1)

# Obtengo solución
sVo1= sym.Eq(Vo1, s_Vo1[0])
sym.pprint(sVo1)

# Fórmula en formato latex
print('Fórmula en formato LaTex: ')
sym.print_latex(sVo1)
sVo1

# Considerando R1= R2= R3= R4= R5= R
s_Vo1= sVo1.subs({R1: R, R2: R, R3: R, R4: R, R5: R})
s_Vo1
```

Se analiza la segunda etapa

```python
# Cargo las variables particulares.
Vx, Vo= sym.symbols('V_{x}, V_{o}')

# Nodo Vx es igual a la salida Vo1
Vx= sVo1.rhs

# Se aplica LCK
Vop= sym.Eq(((Vo2-V2)/R3), ((V2-Vx)/R5)+(V2/R1))

# Se normaliza el valor de las resistencias
s_Vo= Vop.subs({R1: R, R2: R, R3: R, R4: R, R5: R})

# Se despeja Vo.
sVo2= sym.solve(s_Vo, Vo2)

# Se arma la ecuación de Vo
s= sym.Eq(Vo, sVo2[0])

sym.pprint(s)
#sym.print_latex(s)
s
```
