### Carga de variables generales del circuito.

```python
import sympy as sym

V1, V2= sym.symbols('V_{1}, V_{2}')
R1, R2, R3, R4, R5= sym.symbols('R_{1}, R_{2}, R_{3}, R_{4}, R_{5}')
```

### Análisis de $V_{1}\neq0V$ y $V_{2}=0V$

```python
from IPython.display import Image

Image(filename= "img/Circuito 1 1.png") 
```

Se tienen las siguientes ecuaciones:

$$ R_{p}= \frac{R_{1}R_{4}}{(R_{1}+R_{4})} $$

$$ I_{R2}= \frac{V_{o1}-V_{1}}{R_{2}} $$

$$ I_{Rp}= \frac{V_{1}}{R_{p}} $$

Se tiene que:

$$ \boxed{ I_{R2}= I_{Rp} }$$

```python
Rp, Ir2, Irp, Vo1= sym.symbols('R_{p}, I_{R2}, I_{Rp}, V_{o1}')

equ_Ir2= sym.Eq(Ir2, (Vo1-V1)/R2)
equ_Irp= sym.Eq(Irp, V1/Rp)

#sym.pprint(equ_Ir2)
#sym.pprint(equ_Irp)

equ1= sym.Eq(equ_Ir2.rhs, equ_Irp.rhs)
sym.pprint(equ1)

s_Vo1= sym.solve(equ1, Vo1)

sVo1= sym.Eq(Vo1/V1, s_Vo1[0]/V1)
sym.pprint(sVo1)

s_Vo1= sVo1.subs(Rp, (R1*R4)/(R1+R4))
sym.pprint(s_Vo1)

sym.print_latex(s_Vo1)

s_Vo1
```

Se calcula la ganancia $V_{o2}/V_{o1}$. 

$$ I_{R3}= \frac{V_{o2}}{R_{3}} $$

$$ I_{R5}= \frac{-V_{o1}}{R_{5}} $$

```python
Ir3, Ir5= sym.symbols('I_{R3}, I_{R5}')
Vo2= sym.symbols('V_{o2}')

# Definir ecuaciones
equ_Ir3= sym.Eq(Ir3, Vo2/R3)
equ_Ir5= sym.Eq(Ir5, -Vo1/R5)

# Igualar expresiones Ir3 e Ir5
equ2= sym.Eq(equ_Ir3.rhs, equ_Ir5.rhs)

# Resuelvo sistema, se obtiene solución en formato matriz.
s_Vo2= sym.solve(equ2, Vo2)

# Obtengo solución
sVo2= sym.Eq(Vo2/Vo1, s_Vo2[0]/Vo1)
sym.pprint(sVo2)

# Fórmula en formato Latex
sym.print_latex(sVo2)

# Fórmula en formato simbólico
sVo2
```

Ganancia $A_{V1}= \frac{V_{o2}}{V_{o1}} \frac{V_{o1}}{V_{1}}$.

```python
Av1, R, Vo= sym.symbols('A_{V1}, R, V_{o}')

equ_Av1= sym.Eq(Av1, sVo2.rhs*s_Vo1.rhs)

sym.pprint(equ_Av1)
equ_Av1.subs({R1: R, R2: R, R3: R, R4: R, R5: R})

s1= sym.Eq(Vo, -3*V1)

s1
```

### Análisis de $V_{1}=0V$ y $V_{2}\neq0V$

```python
from IPython.display import Image

Image(filename= "img/Circuito 1 2.png") 
```

Se calcula en primer lugar $V_{o1}$ en función de $V_{2}$.

Se tienen las siguientes ecuaciones:

$$ I_{R2}= \frac{V_{o1}}{R_{2}} $$

$$ I_{R1}= \frac{-V_{2}}{R_{1}} $$

Se igualan las corrientes $I_{R2}$ e $I_{R1}$.

$$ \boxed{ I_{R2}=I_{R1} } $$

```python
Ir1, R= sym.symbols('I_{R1}, R')

equ_Ir3= sym.Eq(Ir2, Vo1/R2)
equ_Ir5= sym.Eq(Ir1, -V2/R1)

equ2= sym.Eq(equ_Ir3.rhs, equ_Ir5.rhs)
sym.pprint(equ2)

s_Vo1= sym.solve(equ2, Vo1)

sVo1= sym.Eq(Vo1, s_Vo1[0])
sym.pprint(sVo1)

s_Vo1= sVo1.subs({R1: R, R2: R, R3: R, R4: R, R5: R})
s_Vo1 
```

Se calcula en primer lugar la ganancia $V_{o}$ en función de $V_{2}$. Se tienen las siguientes ecuaciones:

$$ \frac{V_{o2}-V_{2}}{R_{3}}= \frac{V_{2}-V_{x}}{R_{5}} + \frac{V_{2}}{R_{1}} $$

```python
Vx, Vo= sym.symbols('V_{x}, V_{o}')

Vx= sVo1.rhs

Vop= sym.Eq( ((Vo2-V2)/R3), ((V2-Vx)/R5)+(V2/R1) )

s_Vo= Vop.subs({R1: R, R2: R, R3: R, R4: R, R5: R})

sVo2= sym.solve(s_Vo, Vo2)

s2= sym.Eq(Vo, sVo2[0])

sym.pprint(s2)

s2
```

Expresión de salida en función del aporte de cada una de las fuentes $V_{1}$ y $V_{2}$.
